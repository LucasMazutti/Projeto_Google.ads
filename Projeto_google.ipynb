```python
#etapa para padronizar as colunas str. Muitas informações e digitações erradas.

df['Campaign_Name'] = df['Campaign_Name'].str.lower()
df['Campaign_Name'] = df['Campaign_Name'].replace({
    'Data Analytics Corse': 'data analytcis course',
    'Data Analytics Course': 'data analytcis course',
    'Data analytcis course	': 'data analytcis course',
    'dataanalyticscourse' : 'data analytcis course'
})
```

```python
df['Campaign_Name'] = df['Campaign_Name'].str.lower()
df['Campaign_Name'] = df['Campaign_Name'].replace({
    'data anlytics corse' : 'data analytcis course',
    'data analytics corse' : 'data analytcis course'})
```

```python
df['Campaign_Name'] = df['Campaign_Name'].str.lower()
df['Campaign_Name'] = df['Campaign_Name'].replace({
    'data analyticis course' : 'data analytcis course'})
```

```python
df['Campaign_Name'] = df['Campaign_Name'].str.lower()
df['Campaign_Name'] = df['Campaign_Name'].replace({
    'data analytcis course' : 'Data Anlytics Course' })
```

```python
df['Location'] = df['Location'].str.lower()
df['Location'] = df['Location'].replace({
    'hyderabad': 'Hyderabad',
    'hydrebad': 'Hyderabad',
    'hyderbad': 'Hyderabad',
    'HYDERABAD' : 'Hyderabad'
})
```

```python
df['Device'] = df['Device'].str.lower()
```

```python
df.head(10)
```

```python
df['Ad_Date'] = df['Ad_Date'].str.replace('-', '/', regex=False)
```

```python
#vamos separar partes da data, Primeiro passo para padronizar totalmente.
split_date = df['Ad_Date'].str.split('/', expand=True)
```

```python
#checar se a primeira parte tem 4 dígitos (ano no caso).
mask_year_first = split_date[0].str.len() == 4
```

```python
df.loc[mask_year_first, 'Ad_Date'] = (
    split_date.loc[mask_year_first, 2] + '/' +
    split_date.loc[mask_year_first, 1] + '/' +
    split_date.loc[mask_year_first, 0]
)
```

```python
df['Ad_Date'] = pd.to_datetime(df['Ad_Date'], dayfirst=True)
```

```python
print(df['Ad_Date'].head())
# não ficou totalmente como desejado, mas assim ja serve. 
```

```python
#Tratamento de símbolos em colunas numéricas
df['Cost'] = df['Cost'].replace(r'[\$,₹]', '', regex=True).astype(float)
df['Sale_Amount'] = df['Sale_Amount'].replace(r'[\$,₹]', '', regex=True).astype(float)

# Com r'[\$,₹]' o regex entende \$ como “símbolo de dólar” normalmente, sem alerta.
```

```python
print(df.isnull().sum())
```

```python
#tratamento de valores nulos. Não podemos ter uma regra global. Precisa ser analisado coluna por coluna.
#Clicks, Impressions, Cost, Leads, Conversions, Sale_Amount
#Essas colunas são quantitativas / numéricas, ligadas a desempenho.
#preenchiemnto igual a 0 ou para não ter outliers, preenchimento com mediana.
```

```python
#Vamos importar o pacote NUMPY para alguns calculos de confirmação de resulatdo da coluna 'conversion rate'.
import numpy as np
```

```python
cols_to_zero = ['Clicks', 'Impressions', 'Cost', 'Leads', 'Conversions', 'Sale_Amount']
df[cols_to_zero] = df[cols_to_zero].fillna(0)
```

```python
df.head()
```

```python
# Recalcular Conversion Rate
df['Conversion Rate'] = df['Conversions'] / df['Clicks'].replace(0, np.nan)
```

```python
df['Conversion Rate'] = df['Conversion Rate'].round(4)
```

```python
print(df.isnull().sum())
```

```python
#após recalcular, alguns valores manteram 0 por não ter a informação 'conversions' ou 'clicks'.
#vamos preencher com zero.
df['Conversion Rate'] = df['Conversion Rate'].fillna(0)
```
